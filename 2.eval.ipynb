```python
from dataloader import Color_Dataloader
import torch
from eval import eval,plot_confusion_matrix
import pandas as pd


import warnings
warnings.filterwarnings("ignore")
```

```python
DIR_DATASET = 'dataset'
BATCH_SIZE = 10
PATH_MODEL = 'weights/ColorNet_model.pt'
DEVICE = 'cuda'
```

```python
model_simple = torch.load(PATH_MODEL)
```

```python
train_loader, test_loader = Color_Dataloader(
    dir_dataset=DIR_DATASET,
    batch_size=BATCH_SIZE
)
```

```python
predics, labels = eval(model = model_simple, 
    dataloader=train_loader, 
    device = DEVICE)
```

```python
plot_confusion_matrix(labels, predics)
```

```python
predics, labels = eval(model = model_simple, 
    dataloader=test_loader, 
    device = DEVICE)
```

```python
plot_confusion_matrix(labels, predics)
```
